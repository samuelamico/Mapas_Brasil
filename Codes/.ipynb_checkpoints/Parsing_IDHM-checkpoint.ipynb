```python
import pandas as pd
```

```python
file_idhm = open("IDHM Municípios 2010.txt","r",encoding="utf8")
data = file_idhm.read()
```

```python
#### Mapping Parsing

# retirar o head e o final do body
map1 = data.split("</tr>")[1:-1]
map2 = []
for item in map1:
    map2.append( item.split('\n') )
                
map2
```

```python
### Build Dataframe:

def build_dataframe(df,lists):
    names =['id','municipio','IDHM','IDHM Renda','IDHM Longevidade','IDHM Educacao']
    if(len(lists) < 6):
        print(lists)
    for i,iten in enumerate(lists):
        if(i == 6): break
        # Construtct Dataframe:
        if names[i] in df:
            df[names[i]].append(iten)
        else:
            df[names[i]] = [iten]

    return df

df_cidade_idhm = {}

for features in map2:
    feat_list = []
    for feat in features:
        try:
            feat = feat.split('<tr>')[1]
        except:
            pass
        feat_spliter = str(feat.split('</td>')[0]).split('<td>')[-1]
        feat_list.append(feat_spliter)
    
    build_dataframe(df_cidade_idhm,feat_list)
    
    
```

```python
print (len(df_cidade_idhm['id']) , \
len(df_cidade_idhm['municipio']), \
len(df_cidade_idhm['IDHM']), \
len(df_cidade_idhm['IDHM Renda']), \
len(df_cidade_idhm['IDHM Longevidade']), \
len(df_cidade_idhm['IDHM Educacao']), \
      )
```

```python
dataframe_idhm = pd.DataFrame.from_dict(df_cidade_idhm)
```
