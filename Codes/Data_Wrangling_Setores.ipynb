# Adicionando e Ajeitando informações dos GeoJSONs

```python
import json
import pandas as pd
import glob
```

## Adicionand População aos Setores e criando um novo GeoJSON

```python
df_populacao_setores = pd.read_csv(r'C:\Users\samfs\Documents\GITHUB\Mapas_Brasil\Populacao\Distrito_Populacao_2016.csv')
df_populacao_setores.head(10)
```

```python
lista_setores = glob.glob(r'C:\Users\samfs\Documents\GITHUB\Mapas_Brasil\Codes\Setores\*.json')
lista_setores
```

```python
df_setores = df_populacao_setores[df_populacao_setores.Cod_setor == 120001305000001]
df_setores.shape
```

## Atualizar informações para cada Setor

```python
def query(codigo):
    try:
        df_setores = df_populacao_setores[df_populacao_setores.Cod_setor == codigo]
    except:
        return('','','',0)
    
    if(df_setores.shape[0] == 0 ):
        return('','','',0)
    else:
        nome_municipio = df_setores.iloc[0][7]
        cod_municipio = df_setores.iloc[0][8]
        nome_uf = df_setores.iloc[0][3]
        populacao = df_setores.iloc[0][-1]
        return(cod_municipio,nome_municipio,nome_uf,populacao)

def verificar_existencia(feat,name):
    try:
        arg = feat['properties'][name]
    except:
        arg = ''
    return arg

def sigla_ibge(cod):
    dic_estado = {"53":"DF","11":"RO", "12":"AC", "13":"AM", "14":"RR", "15":"PA", "16":"AP", "17":"TO", "21":"MA", "22":"PI" , "23":"CE" , "24":"RN" , "25":"PB" , "26":"PE", "27":"AL" , "28":"SE", "29":"BA", "31":"MG", "32":"ES" , "33":"RJ", "35":"SP", "41":"PR", "42":"SC", "43":"RS", "50":"MS", "51":"MT", "52":"GO"}
    return(dic_estado[cod])


for file in lista_setores:
    # Ler file
    print(file)
    name_file = file.split('\?')[-1]
    json_complete = []
    #print(name_file)
    
    with open(file, encoding='utf-8') as json_file:
        data = json.load(json_file)
        for feat in data['features']:
            # Capturar o id do setor
            try:
                id_setor = feat['properties']['name']
            except:
                id_setor = feat['properties']['Geocodigo_Setor']

            cod_uf = str(id_setor)[0:2]

            ### Func:
            cod_municipio,nome_municipio,nome_uf,populacao = query(int(id_setor))

            
            try:
                geometrie = feat['geometry']['coordinates']
            except:
                geometrie = []
                
            if(len(geometrie) == 0):
                print("nao achou geometria")
                pass
            else:
                if(cod_municipio != '' and nome_municipio != ''):
                    doc = {
                        "type": "Feature",
                        "properties": {
                            "Cod_UF": str(cod_uf),
                            "UF": str(sigla_ibge(cod_uf)),
                            "Cod_Setor": str(id_setor),
                            "Cod_Municipio": str(cod_municipio),
                            "Municipio": str(nome_municipio),
                            "Distrito": verificar_existencia(feat,'Distrito'),
                            "Sub_distrito": verificar_existencia(feat,'Sub_distrito'),
                            "Populacao": str(populacao)
                        },
                        "geometry":{
                            "type": 'Polygon',
                            "coordinates": geometrie
                        }
                    }                    
                else:
                     doc = {
                        "type": "Feature",
                        "properties": {
                            "Cod_UF": str(cod_uf),
                            "UF": str(sigla_ibge(cod_uf)),
                            "Cod_Setor": str(id_setor),
                            "Cod_Municipio": verificar_existencia(feat,'Geocodigo_Município'),
                            "Municipio": verificar_existencia(feat,'Município'),
                            "Distrito": verificar_existencia(feat,'Distrito'),
                            "Sub_distrito": verificar_existencia(feat,'Sub_distrito'),
                            "Populacao": str(populacao)
                        },
                        "geometry":{
                            "type": 'Polygon',
                            "coordinates": geometrie
                        }
                    }
                json_complete.append(doc)
            
    print("Fim")
    ##### Salvar arquivo
    geojson = {
        "type":"FeatureCollection",
        "features": json_complete
    }
    with open(name_file, 'w') as json_file:
        json.dump(geojson, json_file, ensure_ascii=True)
```
